```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
data=pd.read_csv('Zomato dataset.csv')
data.head()
```

```python
data.columns
```

```python
data.info()
```

```python
# CONVERT THE DATA TYPE OF THE RATE COLUMNS TO FLOAT AND REMOVE THE DENOMINATOR.
```

```python
def data_rate(rate):
  if isinstance(rate,str):
    return float(rate.split('/')[0])

  else:
    return rate

data['rate']=data['rate'].apply(data_rate)
```

```python
data.info()
```

```python
data.head()
```

```python
sns.countplot(x= 'listed_in(type)', data=data , palette='viridis')
plt.xlabel("Type of restaurant")
plt.xticks(rotation=90)
plt.show()
```

```python
groupe_daata=data.groupby('listed_in(type)')['votes'].sum().reset_index()
```

```python
plt.figure(figsize=(8,4))
sns.lineplot(x='listed_in(type)',y='votes',data=groupe_daata,marker='*',markersize=20 ,c='green')
plt.xlabel("Type of restaurant")
plt.ylabel("votes")
plt.title("Acoording votes and reutorants ")
plt.xticks(rotation=90)
plt.show()
```
